```python
import os
import csv
import functools
root='./data/5/0728(3125çµ„)/'
all=os.listdir(root)
tot=0
pol=[]
for i in all:
    if(i=='text.txt'):
        continue
    with open(root+i,'r') as csv1:
        rd=csv1.readlines()
        tot+=len(rd)/2
        pol+=[[i.replace('.csv',''),len(rd)/2]]
print(tot)
```

```python
cnt=[[0for i in range(5)]for j in range(5)]
for i in pol:
    pl=[]
    for j in (i[0][1:-1]).split(','):
        pl+=[int(j)]
    for j in range(5):
        cnt[j][pl[j]]+=i[1]
```

```python
for i in cnt:
    for j in i:
        print(round(j*100/tot,2),end='  ')
    print()
```

```python
def per(cur=[]):
    n=5
    if(len(cur)==n):
        return [cur]
    ret=[]
    for i in range(n):
        if(not i in cur):
            ret+=per(cur+[i])
    return ret
p=per()
print(len(p))
```

```python
cnt=dict()
for f in all:
    with open(root+f,'r') as csv1:
        rd=csv.reader(csv1)
        for i in rd:
            if(i==[]):
                continue
            s=[]
            for j in i:
                s+=[float(j)]
            #print('s=',s)
            ss=sorted(s)
           # print('ss=',ss)
            cur=[]
            for j in p:
                if(s==[ss[j[0]],ss[j[1]],ss[j[2]],ss[j[3]],ss[j[4]]]):
                    cur=j
                    break
           # print('cur=',cur)
            p0=[]
            for j in f.replace('.csv','')[1:-1].split(','):
                p0+=[int(j)]
           # print('p0=',p0)
            p1=[-1,-1,-1,-1,-1]
            for j in range(5):
                p1[cur[j]]=cur[p0[j]]
           # print('p1=',p1)
            if(cnt.get(str(p1))==None):
                cnt[str(p1)]=0
            cnt[str(p1)]+=1
          #  print('\n\n')
#for i in cnt.keys():
 #   print(i,':',cnt[i])
```

```python
def cmp(a,b):
    if(a[1]<b[1]):
        return 1
    return -1
c1=[]
for i in cnt.keys():
    c1+=[[i,cnt[i]]]
c1.sort(key=functools.cmp_to_key(cmp))
for i in c1:
    print(i[0],':',round(i[1]*100/tot,3))
```
